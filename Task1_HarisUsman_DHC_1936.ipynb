## Exploring and Visualizing a Simple Dataset

## Objective:
Learn how to load, inspect, and visualize a dataset to understand data trends and distributions.

## Dataset:
Iris Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Dataset Loading
data = sns.load_dataset("iris")
```

## Dataset Overview

Examine the shape, column names, and sample rows of the dataset to understand its structure.


```python
# Dataset shape
print("Shape of dataset:", data.shape)

# Column names
print("\nColumn names:", data.columns.tolist())

# First few rows
data.head()
```

## Dataset Summary Statistics

Use `.info()` to understand the data types and `.describe()` to explore basic statistics for numeric columns.


```python
# Data types
data.info()

# Statistics for numerical columns
data.describe()
```

## Scatter Plot: Visualizing Relationships Between Features

Creating a pairplot to visually analyze the relationship between different numerical features.


```python
sns.pairplot(data, hue="species", palette="Set2")
plt.suptitle("Pairplot of Iris Features by Species", y=1.02)
plt.show()
```

## Histogram: Value Distributions

```python
data.hist(figsize=(10, 8), color='green', edgecolor='black')
plt.suptitle("Histograms of Iris Features", y=1.02)
plt.tight_layout()
plt.show()
```

## Box Plots: Outlier Detection


```python
plt.figure(figsize=(10, 6))
sns.boxplot(data, palette="pastel")
plt.title("Boxplot of Iris Features")
plt.xticks(rotation=45)
plt.show()
```
